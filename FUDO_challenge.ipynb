# Consultas SQL

```python
# Relación entre la duración de los viajes y el modelo de bicicleta.
# [(1405.72, 'FIT'), (1337.66, 'ICONIC')]

# En esta consulta, no afecta los missing values de id estacion destino
res = cursor.execute(
    "SELECT ROUND(AVG(duracion_recorrido), 2) as tiempo_promedio_de_viaje, modelo_bicicleta \
    FROM recorridos \
    GROUP BY modelo_bicicleta;")

res.fetchall()
```

```python
# Relación entre la edad de los usuarios y la cantidad de viajes que realizaron.
# [(10, 14), (38, 15), (108, 16), (297, 17) ...

res = cursor.execute(
    "SELECT COUNT(re.duracion_recorrido) as cantidad_de_viajes, us.edad \
    FROM recorridos as re, usuarios as us \
    WHERE re.id_usuario = us.id_usuario AND \
    re.id_estacion_origen != re.id_estacion_destino \
    GROUP BY us.edad \
    ORDER BY us.edad \
    LIMIT 10;")

res.fetchall()
```

```python
# Relación entre la cantidad de viajes realizados y la fecha de alta de los usuarios.

res = cursor.execute(
    "SELECT COUNT(re.duracion_recorrido) as cantidad_de_viajes, strftime('%m', us.fecha_alta) as mes \
    FROM recorridos as re, usuarios as us \
    WHERE re.id_usuario = us.id_usuario \
    GROUP BY mes \
    ORDER BY mes;")

res.fetchall()
```

```python
# Comunas que más viajes reciben

res = cursor.execute(
    "SELECT est.comuna, COUNT(re.id_estacion_destino) as llegadas_a_comuna \
    FROM recorridos as re, estaciones as est \
    WHERE re.id_estacion_destino = est.id_estacion \
    GROUP BY est.comuna \
    ORDER BY llegadas_a_comuna DESC;")

res.fetchall()
```

```python
# Comunas de las que más viajes parten

res = cursor.execute(
    "SELECT est.comuna, COUNT(re.id_estacion_origen) as partidas_de_comuna \
    FROM recorridos as re, estaciones as est \
    WHERE re.id_estacion_origen = est.id_estacion AND re.id_estacion_destino IS NOT NULL \
    GROUP BY est.comuna \
    ORDER BY partidas_de_comuna DESC;")

res.fetchall()
```

```python
# Usuarios más activos durante 2023.

# NOTA: En esta query se podria aclarar que sean datos de 2023, extrayendo el anio de las fechas de origen y destino
# y pidiendo que sean del anio 2023. Pero como todo el dataset son datos de 2023, no hace falta.
# Salvo que se agreguen datos de otros anios.

res = cursor.execute(
    "SELECT id_usuario, COUNT(id_usuario) as cantidad_de_viajes \
    FROM recorridos \
    GROUP BY id_usuario \
    ORDER BY cantidad_de_viajes DESC \
    LIMIT 10;")

res.fetchall()
```
